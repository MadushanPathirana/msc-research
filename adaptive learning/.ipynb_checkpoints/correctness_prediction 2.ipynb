```python
import pandas as pd
from sklearn.metrics import classification_report
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('C:/research/adaptive learning/test_goal.csv').drop('Unnamed: 0',axis=1)
```

```python
df.sort_values(by=['user_id','start_time'],inplace=True)
```

```python
df.user_id.nunique()
```

```python
df['sequnce']=df.groupby(['user_id']).cumcount()
```

select 80% of the users as train and 20% as test

```python
train_size=int(df.user_id.nunique()*.65)
test_size=df.user_id.nunique()-train_size
```

```python
users=df.user_id.unique().tolist()
test_users=np.random.choice(users,test_size,replace=False)
```

```python
df=df.drop('start_time',axis=1)
```

drop instructions 

```python
df=df[~df['correct'].isna()].reset_index()
```

```python
#users=pd.get_dummies(df['user_id'])*1
atoms=pd.get_dummies(df['atom_id'])*1
atomType=pd.get_dummies(df['atom_type'])*1
learning_objectives=pd.get_dummies(df['learning_objective_name'])*1
```

```python
dataset=pd.concat([atoms,learning_objectives,df[['correct','sequnce','user_id']]],axis=1)
#dataset=pd.concat([atoms,atomType,df[['correctness','sequnce']]],axis=1)
```

```python
dataset['correct']=dataset['correct']*1
```

```python
train_data=dataset[~dataset.user_id.isin(test_users)]
test_data=dataset[dataset.user_id.isin(test_users)]
```

```python
test_list=[]
train_user_temp_filtered_list=[]
for user in users:
   train_user_temp= test_data[test_data.user_id==user]
   sequnce_list=train_user_temp.sequnce.to_list()
   if len(sequnce_list)>1:
      random_question=np.random.choice(sequnce_list[1:],1)[0]
    
      
      train_user_temp_filtered=train_user_temp[train_user_temp.sequnce < random_question]
      train_user_temp_filtered_list.append(train_user_temp_filtered)
      test_list.append(train_user_temp[train_user_temp.sequnce == random_question])


train_user_temp_filtered_list.append(train_data)
test_data_concat=pd.concat(test_list).drop(['user_id'],axis=1)
train_user_temp_filtered_list.append(train_data)
train_data_concat=pd.concat(train_user_temp_filtered_list).drop(['user_id'],axis=1)

X_train,X_test=train_data_concat.loc[:,train_data_concat.columns!='correct'],test_data_concat.loc[:,test_data_concat.columns!='correct']
y_train,y_test=train_data_concat.loc[:,train_data_concat.columns=='correct'],test_data_concat.loc[:,test_data_concat.columns=='correct']
class LogisticRegression(nn.Module):
    def __init__(self, input_dim):
        super(LogisticRegression, self).__init__()
        self.f1 = nn.Linear(input_dim, 50)
        self.f2 = nn.Linear(50,100)
        self.f3 = nn.Linear(100,1)

    def forward(self, x):
        out = torch.sigmoid(self.f1(x))
        out = torch.sigmoid(self.f2(out))
        out = torch.sigmoid(self.f3(out))
        return out

model = LogisticRegression(X_train.shape[1])

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)

num_epochs = 3000
losses = []
losses_test = []
Iterations = []
iter = 0
for epoch in range(num_epochs):
    inputs = torch.from_numpy(X_train.values.astype(np.int64)).float()
    labels = torch.from_numpy(y_train.values.astype(np.int64)).float()

    optimizer.zero_grad()
    outputs = model(inputs)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    iter+=1
    if (epoch+1) % 100 == 0:
        print('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))

with torch.no_grad():
    inputs = torch.from_numpy(X_test.values.astype(np.int64)).float()
    labels = torch.from_numpy(y_test.values.astype(np.int64)).float()

    outputs = model(inputs)
    predicted = (outputs >= 0.5).squeeze().float()
    accuracy = accuracy_score(labels.numpy(), predicted.numpy())

print('Accuracy: {:.2f}%'.format(accuracy*100))
```

```python
print(classification_report(y_true=labels,y_pred=predicted))
```
