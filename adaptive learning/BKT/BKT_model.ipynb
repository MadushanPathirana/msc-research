```python
import numpy as np
import pandas as pd
from pyBKT.models import Model
import matplotlib.pyplot as plt
```

```python
model = Model(seed = 42, num_fits = 1)
model.fit('bkt_data.csv')
```

```python
bkt_data=pd.read_csv('bkt_data.csv')
```

```python

training_rmse = model.evaluate(data = bkt_data)
training_auc = model.evaluate(data_path = 'bkt_data.csv', metric = 'auc')
print("Training RMSE: %f" % training_rmse)
print("Training AUC: %f" % training_auc)
```

```python
model.evaluate(data_path = 'bkt_data.csv', metric = 'accuracy')
```

```python
reds_df = model.predict(data_path = 'bkt_data.csv')
```
