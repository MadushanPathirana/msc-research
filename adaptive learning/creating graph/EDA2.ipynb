```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_pickle('appmath.pkl')
```

```python
df=df[['graph_id', 'user_id', 
       'goal_id',
        'interaction_end_time',
       'learning_objective_name', 'atom_id', 
       'correct', 'time_spent_answering_s', 'time_spent_on_instruction_s',
       'goal_progress',
       'target_status_and_progress', 'prev_concept_narrative',
       'duration_s', 'is_target']].reset_index(drop=True)
```

```python
for col in df.columns:
    if df[col].dtype.name=='category':
        df[col]=df[col].astype(str)
```

```python
def atomClassifier(x):
    if x==None:
        return 'learningMaterial'
    else:
        return 'question'
```

```python
df['atom_type']=df.correct.apply(lambda x:atomClassifier(x) )
```

```python
df['correctBinary']=df.correct*1
df.atom_id=df.atom_id.astype(str)
```

```python
summary=df[~df.correctBinary.isna()].pivot_table(index='goal_id',aggfunc={'learning_objective_name':pd.Series.nunique,'atom_id':pd.Series.nunique,'user_id':pd.Series.nunique,'interaction_end_time':np.count_nonzero})
```

```python
summary.sort_values(by=['interaction_end_time','atom_id'],ascending=[False,True]).head(20)
```

```python
summary['atom_usage']=summary.interaction_end_time/summary.atom_id
```

```python
summary.sort_values('atom_usage',ascending=False).head(10)
```
