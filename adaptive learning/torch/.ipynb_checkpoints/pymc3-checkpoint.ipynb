```python
import numpy as np
import pymc3 as pm

# Generate some sample data
np.random.seed(123)
N = 100
sigma = 0.5
beta0 = 1
beta1 = 2
x = np.linspace(-1, 1, N)
y = beta0 + beta1*x + np.random.normal(scale=sigma, size=N)

# Define the model
with pm.Model() as model:
    # Define the prior distribution over the regression coefficients
    beta0 = pm.Normal('beta0', mu=0, sd=10)
    beta1 = pm.Normal('beta1', mu=0, sd=10)

    # Define the likelihood function
    mu = beta0 + beta1*x
    likelihood = pm.Normal('y', mu=mu, sd=sigma, observed=y)

    # Run the inference algorithm
    trace = pm.sample(10, tune=10)

# Print a summary of the posterior distribution over the regression coefficients
print(pm.summary(trace))
```

```python
import theano.tensor as tt
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simulate data from a known mixture distribution
np.random.seed(12345) # set random seed for reproducibility
data = np.load('vals.npy')
k = 3
ndata = data.shape[0]
spread = .5
centers = np.array([-spread, 0, spread])

# simulate data from mixture distribution
v = np.random.randint(0, k, ndata)


plt.hist(data);
```

```python
# setup model
model = pm.Model()
with model:
    # cluster sizes
    p = pm.Dirichlet('p', a=np.array([1., 1., 1.]), shape=k)
    # ensure all clusters have some points
    p_min_potential = pm.Potential('p_min_potential', tt.switch(tt.min(p) < .1, -np.inf, 0))


    # cluster centers
    means = pm.Normal('means', mu=[0, 0, 0], sd=15, shape=k)
    # break symmetry
    order_means_potential = pm.Potential('order_means_potential',
                                         tt.switch(means[1]-means[0] < 0, -np.inf, 0)
                                         + tt.switch(means[2]-means[1] < 0, -np.inf, 0))

    # measurement error
    sd = pm.Uniform('sd', lower=0, upper=20)

    # latent cluster of each observation
    category = pm.Categorical('category',
                              p=p,
                              shape=ndata)

    # likelihood for each observed value
    points = pm.Normal('obs',
                       mu=means[category],
                       sd=sd,
                       observed=data)
```

```python
# fit model
with model:
    step1 = pm.Metropolis(vars=[p, sd, means])
    step2 = pm.ElemwiseCategorical(vars=[category], values=[0, 1, 2])
    tr = pm.sample(1000, step=[step1, step2])
```

```python
pm.plots.traceplot(tr, ['p', 'sd', 'means']);
```

```python
pm.summary(tr)
```

```python
posterior_probs = pm.sample_posterior_predictive(tr, model=model)['obs']

# assign each data point to the cluster with the highest posterior probability
data_clusters = []
for data_arr in [np.array([.2,.3,.4]), np.array([.4,.2,.5,.8,1])]:
    cluster_probs = np.mean(posterior_probs[:, data_arr.shape[0]:], axis=1)
    data_cluster = np.argmax(cluster_probs)
    data_clusters.append(data_cluster)

print(data_clusters)  # prints [0, 1]
```
