```python
subgraphs=[G.subgraph(i) for i in nx.connected_components(G)]
```

```python
for num,sub in enumerate(subgraphs):
    
    nx.set_node_attributes(sub,num,'subGraphId')
```

```python
nodeSubGraphId={}
for g in subgraphs:
    for node in g.nodes(data=True):
         nodeSubGraphId[node[0]]= node[1]['subGraphId']
```

```python
progress=sampleUser.pivot_table(index='learning_objective_name',values='goal_progress',aggfunc=np.max).reset_index().rename(columns={'learning_objective_name':'source_lo_title'})
```

```python
progressMap=progress.merge(df[['source_lo_title','dest_lo_title']], on='source_lo_title',how='left')
```

```python
progressMap[progressMap.source_lo_title=='Determine if two switching circuits are equivalent']
```

```python
progressMap.dest_lo_title=progressMap.dest_lo_title.fillna(0)
```

```python
G2=nx.Graph()

for _,i in progressMap.iterrows():   
   
    if i['dest_lo_title']==0:
        G2.add_node(i['source_lo_title'],color=i['goal_progress'])

    else:
       
        G2.add_edge(i['source_lo_title'],i['dest_lo_title'])
        G2.add_node(i['source_lo_title'],color=i['goal_progress'])
     
```

```python

nodes = G2.nodes()

for n in nodes:
       if G2.nodes[n].get('color')==None:
            G2.nodes[n]['color']=-1

```

```python
colorMap='PiYG'
cmap=  plt.cm.get_cmap(colorMap)
```

```python
def Colors(c):
    if c==-1:
        return (0,0,0)
    else:
       return cmap(c)
```

```python

node_colors = [Colors(G2.nodes[n]['color']) for n in G2.nodes]
fig = plt.figure(1, figsize=(20, 20), dpi=100) 
# plot the graph
pos = nx.spring_layout(G2)
nx.draw(G2, pos, node_color=node_colors,node_size=25,font_size=7,with_labels=False)

# create a colorbar legend
sm = plt.cm.ScalarMappable(cmap=plt.cm.get_cmap(colorMap), norm=plt.Normalize(vmin=0, vmax=1))
sm._A = []
plt.colorbar(sm)

# show the plot
plt.show()
```

```python
sampleUser2=sampleUser.sort_values(by='interaction_end_time')[['learning_objective_name','correct','atom_id']]
```

```python
sampleUser2.learning_objective_name=sampleUser2.learning_objective_name.astype(str)
sampleUser2.atom_id=sampleUser2.atom_id.astype(str)
```

```python
sampleUser2['postAtom']=sampleUser2.atom_id.shift(-1)
```

```python
sampleUser2=sampleUser2[~sampleUser2.postAtom.isna()]
```

```python
sampleUser2.correct=sampleUser2.correct.fillna('0')
```

```python
loList=sampleUser2.learning_objective_name.unique()
```

```python
sampleUser2.correct=sampleUser2.correct.astype(str)
```

```python
sampleUser2.correct=sampleUser2.correct.replace('False','2')
sampleUser2.correct=sampleUser2.correct.replace('True','1')
```

```python
colorDic={'1':'green','2':'red','0':'yellow'}
```

```python
sampleUser2
```

```python
G3=nx.Graph()

for i in loList:
    sampleUser2Temp=sampleUser2[sampleUser2.learning_objective_name==i]
    G3.add_edge(i,sampleUser2Temp.atom_id.values[0])
    G3.add_node(i,color='blue')
    for num, row in sampleUser2Temp.iterrows():
        G3.add_edge(row['atom_id'],row['postAtom'])
        G3.add_node(row['atom_id'],color=colorDic[row['correct']])
        
    for num2,row2 in df[df.source_lo_id==i].iterrows():
        G3.add_edge(row2['source_lo_title'],row2['dest_lo_title'])
        G3.add_node(row2['dest_lo_title'],color='blue')
        

```

```python
for n in  G3.nodes(data=True):
   if  n[1]=={}:
      n[1]['color']=  colorDic[sampleUser2[sampleUser2.postAtom==n[0]].correct.values[0]]
```

```python

    
node_colors = [G3.nodes[n]['color'] for n in G3.nodes]
fig = plt.figure(1, figsize=(20, 20), dpi=100) 
# plot the graph
pos = nx.circular_layout(G3)
nx.draw(G3, pos,node_color=node_colors, node_size=25,font_size=7,with_labels=False)

# create a colorbar legend
#sm = plt.cm.ScalarMappable(cmap=plt.cm.get_cmap(colorMap), norm=plt.Normalize(vmin=0, vmax=1))
#sm._A = []
#plt.colorbar(sm)

# show the plot
plt.show()
```
